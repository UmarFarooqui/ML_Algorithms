```python
boston = load_boston()
```

```python
bos = pd.DataFrame(boston.data)
bos.head()
```

```python
# Add the column names
bos.columns = boston.feature_names
bos.head()
```

```python
# Add the target column to the dataframe
bos['PRICE'] = boston.target
bos.head()
```

```python
# Summary statistics
bos.describe()
```

```python
# EDA
plt.scatter(bos.CRIM, bos.PRICE)
plt.xlabel("Per capita crime rate by town (CRIM)")
plt.ylabel("Housing Price")
plt.title("Relationship between CRIM and Price")
```

```python
# Scatter plot between RM and PRICE
plt.scatter(bos.RM, bos.PRICE)
plt.xlabel("Average number of rooms per dwelling (RM)")
plt.ylabel("Housing Price")
plt.title("Relationship between RM and Price")
```

```python
# Kind of relationship
sns.regplot(y="PRICE", x="RM", data=bos, fit_reg = True)
# The relationship between RM and PRICE is positive and linear
# We see a small number of outliers also
# RM seems to be a better predictor for the housing price
```

```python
# Scatter plot between PTRATIO and PRICE
plt.scatter(bos.PTRATIO, bos.PRICE)
plt.xlabel("Pupil-teacher ratio by town (PTRATIO)")
plt.ylabel("Housing Price")
plt.title("Relationship between PTRATIO and Price")
```

```python
# Scatter plot between DIS and PRICE
plt.scatter(bos.DIS, bos.PRICE)
plt.xlabel("Weighted distances to five Boston employment centres (DIS)")
plt.ylabel("Housing Price")
plt.title("Relationship between DIS and Price")
```

```python
# Kind of relationship
sns.regplot(y="PRICE", x="DIS", data=bos, fit_reg = True)
# The relationship is positive and linear
# Also, we see a considerable number of outliers
```

```python
# Fitting linear regression using statsmodels
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
type(bos)
```

```python
m = ols('PRICE ~ RM', bos).fit()
```

```python
print(m.summary())
```

```python
# Scatterplot between the predicted prices and the original prices
plt.scatter(m.fittedvalues, bos.PRICE)
plt.xlabel("Predicted Housing Price")
plt.ylabel("Original Housing Price")
plt.title("Relationship between Predicted Housing Price and Original Housing Price")
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
X = bos.drop('PRICE', axis=1)
cls = LinearRegression()
cls.fit(X, bos.PRICE)
predictions = cls.predict(X)
```

```python
# Scatterplot between predicted prices and original prices using the linear regression
plt.scatter(predictions, bos.PRICE)
plt.xlabel("Predicted Housing Price")
plt.ylabel("Original Housing Price")
plt.title("Relationship between Predicted Housing Price and Original Housing Price")
```

```python
print("Estimated intercept coefficient: {}".format(cls.intercept_))
```

```python
print("Number of coefficients: {}".format(len(cls.coef_)))
```

```python
# Showing the coefficients per feature
pd.DataFrame({'Features': X.columns, 'Estimated Coefficients': cls.coef_})
```

```python
# Lets change the model to not fit an intercept term
cls_wo_intercept = LinearRegression(fit_intercept=False)
cls_wo_intercept.fit(X, bos.PRICE)
predictions_wo_intercept = cls_wo_intercept.predict(X)
```

```python
# Scatterplot between predicted prices and original prices using the linear regression
plt.scatter(predictions_wo_intercept, bos.PRICE)
plt.xlabel("Predicted Housing Price without Intercept")
plt.ylabel("Original Housing Price")
plt.title("Relationship between Predicted Housing Price without Intercept and Original Housing Price")
# Without intercept, the regression line goes through the origin
# This might not make sense for some of the features to have the value 0 when the housing price is 0
# e.g. if CRIM is 0, which means that there is no crime in the town, it would mean a higher housing price if we keep all the other features same.
# It is acceptable to not have intercept when we are certain about the true relationship between feature variable and the output variable.
```

```python
print("Estimated intercept coefficient: {}".format(cls_wo_intercept.intercept_))
```

```python
print("Number of coefficients: {}".format(len(cls_wo_intercept.coef_)))
```

```python
pd.DataFrame({'Features': X.columns, 'Estimated Coefficients': cls_wo_intercept.coef_})
```
